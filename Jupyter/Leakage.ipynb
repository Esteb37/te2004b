# Frequency leakage

```python
import numpy as np
from numpy import sin, cos, pi
import matplotlib.pyplot as plt
from matplotlib.pyplot import subplots
from scipy.fftpack import fft, ifft, fftshift


fs = 64
ts = 1/fs
f0_1 = 3
f0_2 = 4
f0_3 = 5
f0_4 = 6

N = np.arange(fs)
x_1 = sin(2*pi*f0_1*N*ts)
x_2 = sin(2*pi*f0_2*N*ts)
x_3 = sin(2*pi*f0_3*N*ts)
x_4 = sin(2*pi*f0_4*N*ts)

plots = lambda: plt.subplots(figsize=(20,10))

f, ax = plots()

ax.plot(N, x_1, 'r-',label='x_1')
ax.plot(N, x_2, 'b-',label='x_2')
ax.plot(N, x_3, 'g-',label='x_3')
ax.plot(N, x_4, 'y-',label='x_4')

plt.show()


X_1 = fft(x_1)
f, ax = plots()
ax.plot(np.arange(-len(X_1)/2, len(X_1)/2), fftshift(abs(X_1)), 'ro', markersize=10,label='X_1')
ax.set_xlim(-5,5)
plt.show()
```

#### What if the number of cycles is not an integer?

```python
fs=64
ts=1/fs
fo_1=3.4
fo_2=4
fo_3=5
fo_4=6

N=np.arange(fs)
x_1=sin(2*pi*fo_1*N*ts)
x_2=sin(2*pi*fo_2*N*ts)
x_3=sin(2*pi*fo_3*N*ts)
x_4=sin(2*pi*fo_4*N*ts)


f, ax=plots()
ax.plot(N, x_1, 'ro')
ax.plot(N, x_2, 'b')
#ax.plot(N, x_3, 'g')
#ax.plot(N, x_4, 'y')

plt.show()

X_1 = fft(x_1)
f, ax = plots()
ax.plot(np.arange(-len(X_1)/2, len(X_1)/2),
        fftshift(abs(X_1)), 'r-', markersize=10, label='X_1')
#ax.set_xlim(-5, 5)
plt.show()
```

```python
fs = 64
ts = 1/fs
f0_1 = 3
N = np.arange(fs)
x_1 = sin(2*pi*f0_1*N*ts)

f, ax = plots()
N = np.arange(-32,32)
x_sinc = 32*np.sinc(pi*f0_1*(3-N)*ts)
X_sinc = fft(x_sinc)
plt.plot(x_sinc)
plt.show()
```
